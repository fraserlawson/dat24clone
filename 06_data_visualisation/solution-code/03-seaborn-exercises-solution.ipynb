### The Data

Loading the Titanic dataset (it is actually included in the Seaborn package!):

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### Exercises

** Recreate the plots below using the Titanic dataframe. **

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell below that one which won't overwrite that plot!* **

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'],bins=30,kde=False,color='red')
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
