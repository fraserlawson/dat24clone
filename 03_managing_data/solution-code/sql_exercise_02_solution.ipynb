# SQL Exercise 02: Rossmann Pharmacy Sales Data

```python
import pandas as pd
import sqlite3
```

> ##### Load the Rossmann sales data and create a table into the database from it

```python
conn = sqlite3.connect('../assets/db/dat-rossman.db')
```

```python
data = pd.read_csv('../assets/data/rossmann_sales.csv', low_memory=False)
data.head()
```

```python
data.to_sql('rossmann_sales',
            con=conn,
            if_exists='replace',
            index=False)
```

```python
pd.read_sql('select * from rossmann_sales limit 10', con=conn)
```

> #### Load the Rossmann Store metadata in `rossmann_stores.csv` and create a table into the database from it

```python
rossmann_stores = pd.read_csv('../assets/data/rossmann_stores.csv')
rossmann_stores.to_sql('rossmann_stores', if_exists='replace', index=False, con=conn)
```

> #### Write a query that returns the `Store`, `Date` and `Customers`

```python
pd.read_sql("""SELECT Store, Date, Customers FROM rossmann_sales;""", con=conn).head()
```

> #### Write a query that returns the `Store`, `Date` and `Customers` for when the stores were open and running a promotion

```python
pd.read_sql("""SELECT Store, Date, Customers FROM rossmann_sales WHERE Open = 1 and Promo = 1;""", con=conn).head()
```

> #### Write a query that returns the total sales on promotion days.

```python
pd.read_sql(
"""
SELECT
Promo, Store, SUM(Sales)
FROM rossmann_sales
GROUP BY Promo
""", con=conn).head()
```
